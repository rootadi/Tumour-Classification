```python
# slicing the loaded data into training and testing data 
# Training data = 75%
# Testing data = 25%
iData=Rawdata.iloc[0:426,1:3]
```

```python
# selecting the M type of tumour 
Mdata=iData[iData['diagnosis'] == 'M']
```

```python
Mdata
```

# Using Bayes Rule
$$ p(class=B/tumour) = \frac{p(tumour/class=B) * p(class=B)}{p(tumour/class=B) * p(class=B) + p(tumour/class=M) * p(class=M)} $$

# Using Gaussian distribution function
$$p(tumour/class=B)) = \frac {1}{\sqrt{2\pi}\sigma_B} \mathrm{e}^{-(\frac{x-\mu_B}{\sqrt{2}\sigma_B})^2} $$

$$p(tumour/class=M)) = \frac {1}{\sqrt{2\pi}\sigma_M} \mathrm{e}^{-(\frac{x-\mu_M}{\sqrt{2}\sigma_M})^2} $$

##### Here the value of MuBCap is as
$$MuBCap = \mu_B $$
    

```python
# MuBCap = p(class = B)  
MuBCap=Bdata['radius_mean'].mean()
```

```python
MuBCap
```

##### Here the value of SigmaBCap is as 
$$ SigmaBCap = \sigma_B $$

```python
SigmaBCap = Bdata['radius_mean'].std()
```

```python
SigmaBCap
```

##### Here the value of MuMCap is as
$$MuMCap = \mu_M $$

```python
MuMCap = Mdata['radius_mean'].mean()
```

```python
MuMCap
```

##### Here the value of SigmaMCap is as 
$$ SigmaMCap = \sigma_M $$

```python
SigmaMCap = Bdata['radius_mean'].std()
```

```python
SigmaMCap
```

##### Here the value of PCapB is as 
$$ PCapB = p(class = B) $$

```python
# PCapB = p(class = B)
PCapB = Bdata.shape[0]/iData.shape[0]
```

```python
PCapB
```

##### Here the value of PCapM is as 
$$ PCapM = p(class = M) $$

```python
PCapM = Mdata.shape[0]/iData.shape[0]
```

```python
PCapM
```

```python
from scipy.stats import norm
```

```python
tData=Rawdata.iloc[426:,1:3]
```

```python
tData
```

```python
tData['radius_mean']
```

```python
Stest = tData['radius_mean']
```

```python
Stest
```

##### Here the value of x is as 
$$ x = P(tumour/class = B) $$

```python
x=norm.pdf(Stest,MuBCap,SigmaBCap)
```

```python
Numerator=x*PCapB
```

##### Here the value of y is as 
$$ y = p(tumour/class=M) $$

```python
y=norm.pdf(Stest,MuMCap,SigmaMCap)
```

```python
z=y*PCapM
```

$$ Denominator = Numerator + p(tumour/class=M)*p(class=M) $$

```python
Denominator=Numerator+z
```

```python
# Here PtumorisBStest = p(class=B/tumour)
PtumorisBStest=Numerator/Denominator
```

```python
PtumorisBStest
```

```python
Tru,Fal=0,0
for i in range(0,143):
    if(PtumorisBStest[i]>0.5):
        Tru+=1
    else:
        Fal+=1
```

```python
Tru
```

```python
Fal
```

```python
d={'PtumorisBStest':PtumorisBStest}
df = pd.DataFrame(d)
```

```python
df
```

```python
df['PtumorisBStest'] = np.where(df['PtumorisBStest']>0.5,'B','M')
```

```python
df
```

```python
tData = tData.reset_index(drop=True)
```

```python
tData
```

```python
result=pd.concat([df,tData],axis=1)
```

```python
result
```

```python
# Here we are making true positive, true negative, false positive and false negative 
conditions = [
    (result['PtumorisBStest']=='B') & (result['diagnosis']=='B'),
    (result['PtumorisBStest']=='B') & (result['diagnosis']!='B'),
    (result['PtumorisBStest']=='M') & (result['diagnosis']=='M'),
    (result['PtumorisBStest']=='M') & (result['diagnosis']!='M')
]
choices=['true_positive','true_negative','false_positive','false_negative']
result['eval'] = np.select(conditions, choices)
```

```python
result
```
